```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)

data = {
    'Emp_ID': range(1001, 1016),
    'Name': ['Shubham','Suraj','Unmesh','Ganesh','Amit','Rohit','Kiran','Suresh',
             'Neha','Pooja','Rahul','Ankit','Vikas','Sneha','Ravi'],
    
    'Age': np.random.randint(22, 40, 15),
    
    'City': np.random.choice(['Pune','Mumbai','Nashik','Bangalore'], 15),
    
    'Department': np.random.choice(['IT','HR','Finance','Sales'], 15),
    
    'Salary': np.random.randint(4000, 25000, 15),
    
    'Experience': np.random.randint(1, 10, 15),
    
    'Joining_Year': np.random.randint(2015, 2024, 15),
    
    'Performance_Score': np.random.randint(1, 6, 15),
    
    'Bonus': np.random.randint(500, 5000, 15)
}

df = pd.DataFrame(data)
```
